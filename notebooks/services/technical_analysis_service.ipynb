```python
from config.settings import Settings
from integration.clients.market_data.massive_data_client import MassiveDataClient
from integration.clients.market_data.yfinance_data_client import YFinanceDataClient
from integration.providers.market_data.live_data_provider import LiveDataProvider

from application.services.technical.technical_analysis_service import (
    TechnicalAnalysisService,
)
from application.services.base import ServiceRequest
from application.services.base import ServiceRunner
from application.services.technical.technical_request import (
    TechnicalAnalysisRequest,
)

settings = Settings()
massive_client = MassiveDataClient(settings=settings)
yf_client = YFinanceDataClient()

provider = LiveDataProvider(
    massive_data_client=massive_client, yfinance_data_client=yf_client, telemetry=None
)

technical_service = TechnicalAnalysisService(data_provider=provider, event_bus=None)

service_runner = ServiceRunner()

technical_result = await service_runner.run(
    service=technical_service,
    request=ServiceRequest(
        operation="technical.analysis.execute",
        payload=TechnicalAnalysisRequest(
            symbol="SPY",
            days=365,
        ),
    ),
)

technical_result.raise_if_failed()
if technical_result.data is None:
    raise RuntimeError("Technical service returned no result data.")

tech_snapshot = technical_result.data.analysis
```
